```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
ratings = pd.read_csv("ml-latest-small/ratings.csv")
moviesdf = pd.read_csv("ml-latest-small/movies.csv")
```

```python
ratings = ratings.drop(["timestamp"],axis=1)
ratings.head()
```

```python
moviesdf.head()
```

```python
ratings = ratings.pivot_table(index=["userId"], columns=["movieId"], values="rating")
ratings = ratings.fillna(0)
ratings.head()
```

```python
user1 = ratings.loc[1]
# user1
```

```python
simmilaruser = ratings.transpose().corrwith(user1)
```

```python
sortedusers = [sorted(simmilaruser.to_dict().items(), key = lambda kv:(kv[1], kv[0]))]
```
