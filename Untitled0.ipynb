```python
movies = pd.read_csv("ml-latest-small/movies.csv")
```

```python
movies["lis"] = movies.apply(lambda r: list(r.genres.split("|")),axis=1)
movies.head()
```

```python
mmap = {1: "Action",
2: "Adventure",
3: "Animation",
4: "Children",
5: "Comedy",
6: "Crime",
7: "Documentary",
8: "Drama",
9: "Fantasy",
10: "Film-Noir",
11: "Horror",
12: "Musical",
13: "Mystery",
14: "Romance",
15: "Sci-Fi",
16: "Thriller",
17: "War",
18: "Western",
19: "(no genres listed)"}
mmap = {value:key for key, value in mmap.items()}
```

```python
moviemap = {}
keysofmap = mmap.keys()
for ind in movies.index:
    moviemap[int(movies["movieId"][ind])] = [mmap[movie] for movie in movies["lis"][ind] if movie in keysofmap]
moviemap[3]
```
